### [2.2: Проверим тип данных в столбцах](#content)

Давайте получше изучим столбцы, а именно ту информацию которую они хранять и изучим их тип данных, например ранее отметили, что в столбце `year_of_release` который хранит информацию о дате, можем изменить тип на **datetime**. Изучим оставшиеся столбцы и изменим их тип данных, если потребуется.

```python
# выведем 10 строк таблицы на экран
# выведим тип столбцов
# изменим тип в столбцах, если потребуется

print('Первые 10 строк таблицы:')
print()
display(games.head(10))

print()
print('Данные о типах столбцов:')
games.dtypes
```

Изучив тип данных в столбцах, я предлагаю сделать следующее:
* в следующих столбцах данные записать нижним регистров:
 * `name`;
 * `platform`;
 * `genre`.
* в столбце `year_of_release` изменить тип данных на **datetime**;
* столбцы с типом данных **float64** заменим на **float32** для экономии места.

```python
# приведем данные к нижнему регистру
# заменим тип данных в столбцах

games['name'] = games['name'].str.lower()
games['platform'] = games['platform'].str.lower()
games['genre'] = games['genre'].str.lower()

games['year_of_release'] = pd.to_datetime(games['year_of_release'], format='%Y')


games[['na_sales', 'eu_sales', 'jp_sales', 'other_sales', 'critic_score']] = (
    
    games[['na_sales', 
           'eu_sales', 
           'jp_sales', 
           'other_sales', 
           'critic_score']]
    .astype('float32')

)

```

В процессе мы обратили внимание, что столбец `user_score` - оценка пользователей (максимум 10), имеет тип данных **object**, а хранит числовые значения. Давайте подробнее разберем данный столбец и выведем уникальные значения на экран.

```python
# выведем уникальные значения столбца user_score


print('Уникальные значения столбца user_score:')
print()
games['user_score'].sort_values().unique()
```

Среди числовых значений о рейтинге мы видим значение - **tbd**.

[В статье рейтинг "tbd"](https://fb-ru.turbopages.org/fb.ru/s/article/171345/tbd---chto-znachit-eto-sokraschenie) - это "To Be Determined", то есть **"Будет определено"**. В нашем случае думаю, что рейтинг от пользователь еще не определен, давайте подробнее взглянем на все игры с таким значением.

```python
# выведем на экран игры с рейтингом tbd

print('Количество игр с рейтингом "tbd": {0}'.format(games.query('user_score == "tbd"')['name'].nunique()))
print()
print('Таблица с играми у которых рейтинг "tbd":')
games.query('user_score == "tbd"')
```

**1 941** уникальной игры с данным рейтингом, а всего игр с таким рейтингом на разных платформах равняется **2 424**.

Напомню, что столбец `user_score` - это рейтинг к игре, который проставили пользователи, и он рассчитывается в диапазоне от **0** до **10** баллов, и это категориальные значения, мы не можем изменить здесь рейтинг на **0** - это будет означать не пропуск рейтинга а очень плохую оценку игре от пользователей. Поэтому предлагаю значение **tbd** превратить в пропущенные значения типа *nan* и будем считать, что рейтинг к этой игре попросту не ставили.

И еще после изменим тип данных в столбце `user_score` на **float32**.

```python
# заменим tbd на NaN и изменим тип данных на float32 

games = games.replace({'user_score':{'tbd': np.nan}})
games['user_score'] = games['user_score'].astype('float32')
```

ESRB была основана в 1994 году ассоциацией Interactive Digital Software Association (современное название — «Entertainment Software Association»). К началу 2003 года рейтинги ESRB получили более 8000 продуктов от 350 издателей.

Виды рейтингов **ESRB**:
* **«EC»** («Early childhood») — «Для детей младшего возраста;
* **«E»** («Everyone») — «Для всех»;
* **«E10+»** («Everyone 10 and older») — «Для всех от 10 лет и старше»;
* **«T»** («Teen») — «Подросткам»;
* **«M»** («Mature») — «Для взрослых»;
* **«AO»** («Adults Only 18+») — «Только для взрослых»;
* **«RP»** («Rating Pending») — «Рейтинг ожидается».

Теперь виды рейтинга нам известны, давайте взглянем на уникальные значения в этом столбце и сравним виды из столбца с видами выше, тем самым проверим все ли корректно указано у нас в таблице.

### [3.2: Исследуем продажи игр по платформам](#content)

Построим столбчатую диаграмму и изучим на какие платформы было выпущенно больше всего игр в нашей выборке.

```python
# построим столбчатую диаграмму и узнаем на какую платформу было выпущенно больше всего игр

(
    games
    .query('year_of_release > 2012')
    .pivot_table(
        index='platform', 
        values='all_sales', 
        aggfunc='count')
    
    .reset_index()
    
    .sort_values(by='all_sales')
    
    .plot(kind='barh',
         x='platform',
         y='all_sales',
         figsize=(12,8),
         grid=True)
)
plt.title('Количество выпущенных игр')
plt.xlabel('количество (шт.)')
plt.ylabel('игровые платформы');
```

Давайте исследуем **жизнь** каждой платформы из нашего ТОП-списка - как долго она поддерживалась до прихода следующего поколения. Для этого построим столбчатую диаграмму и выведем среднее значение для всех платформ, кроме pc.

```python
# рассчитаем среднею продолжительнность жизни платформы до появления следующего поколения


life_of_platforms = (
    games
    .query('platform != "pc"')
    .pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max'])
)
life_of_platforms.columns = ['min', 'max']
life_of_platforms['duration'] = life_of_platforms['max'] - life_of_platforms['min']
print('''

Средняя продолжительнность выпуска платформы составляет: {0}'''
      .format(life_of_platforms['duration'].median()) 
     
     )
print()
print('Таблица продолжительнности жизни у платформ:')
life_of_platforms.sort_values(by='duration', ascending=False)
```

Таким образом мы можем сказать, что средняя продолжительнность **жизни** игровой приставки составляет **2 191** день или **6 лет!**

### [3.7: Изучим влияние рейтинга ESRB на продажи](#content)

В наборе данных попадается аббревиатура **ESRB** (Entertainment Software Rating
Board) — это ассоциация, определяющая возрастной рейтинг компьютерных
игр. ESRB оценивает игровой контент и присваивает ему подходящую **возрастную
категорию**:

* **«EC»** («Early childhood») — «Для детей младшего возраста;
* **«E»** («Everyone») — «Для всех»;
* **«E10+»** («Everyone 10 and older») — «Для всех от 10 лет и старше»;
* **«T»** («Teen») — «Подросткам»;
* **«M»** («Mature») — «Для взрослых»;
* **«AO»** («Adults Only 18+») — «Только для взрослых»;
* **«RP»** («Rating Pending») — «Рейтинг ожидается».

Но перед этим у нас есть пропущенные значение в этом столбце, которые мы не стали заполнять, давайте эти пропущенные значение заполним значением - **no_rating**, что будет означать, что рейтинга нет.

Для начала давайте посмотрим сколько пропущенных значений у нас присутствует в нашем срезе таблицы **игры от 2013 года выхода**.


```python
# проверим количество пропущенных значений в срезе от 2013 года

print('Количество пропущенных значений в срезе от 2013 года:')

(
games
    .query('year_of_release > 2012')
    .isna()
    .sum()


)
```

Давайте заполни **977** пропущенных значений в столбце `esrb_rating ` на значений **no_rating**. После этого мы построим столбчатую диаграмму между средних мировых продаж к рейтингу и узнаем влияние этого фактора.

```python
# заполним пропущенные значения в столбце esrb_rating на значение - no_rating
# построим столбчатую диаграмму между средних мировых продаж к esrb-рейтингу

games['esrb_rating'] = games['esrb_rating'].fillna('no_rating')

(
    games
    .query('year_of_release > 2012')
    .pivot_table(
        index='esrb_rating', 
        values='all_sales', 
        aggfunc='mean')
    .plot(
        kind='bar', 
        title='Мировые продажи по рейтингу ESRB', 
        xlabel='тип рейтинга ESRB', 
        ylabel='количество проданных копий (млн.)', 
        figsize=(8, 4), 
        grid=True)
    
    
);
```

Видим, что больше всего продаж у возрастной категории:
* **«M»** («Mature») — «Для взрослых».

Меньшевсего продаж у категории:
* **no-rating** - это наше обозначение игр без **возрастного рейтинга**.

Таким образом можем сказать, что по данной выборке меньше всего продаж у игр **без рейтинга ESRB**, следовательнно наличие рейтинга влияет на продажи.

* Европа - создадим датафрейм **eu_users**;
* Япония - создадим датафрейм **jp_users**;
* другие регионы - создадим датафрейм **other_users**.

Затем мы попытаемся составить портрет пользователя каждого из регионов, для этого мы изучим их предпочтения и поведение:

```python
# создадим датафреймы по регионам
# для этого сделаем срез оригинального дф по 2 критериям: год от 2013 (Актуальный период) и продажи чтобы в этом регионе > 0
# удалим лишние столбцы

na_users = games.query('year_of_release > 2012 and na_sales > 0')
na_users = na_users.drop(columns=['eu_sales', 'jp_sales', 'other_sales', 'all_sales'])
na_users.columns = ['name', 
                     'platform', 
                     'year_of_release', 
                     'genre', 
                     'sales',
                     'critic_score', 
                     'user_score', 
                     'esrb_rating']

eu_users = games.query('year_of_release > 2012 and eu_sales > 0')
eu_users = eu_users.drop(columns=['na_sales', 'jp_sales', 'other_sales', 'all_sales'])
eu_users.columns = ['name', 
                     'platform', 
                     'year_of_release', 
                     'genre', 
                     'sales',
                     'critic_score', 
                     'user_score', 
                     'esrb_rating']

jp_users = games.query('year_of_release > 2012 and jp_sales > 0')
jp_users = jp_users.drop(columns=['na_sales', 'eu_sales', 'other_sales', 'all_sales'])
jp_users.columns = ['name', 
                     'platform', 
                     'year_of_release', 
                     'genre', 
                     'sales',
                     'critic_score', 
                     'user_score', 
                     'esrb_rating']

other_users = games.query('year_of_release > 2012 and other_sales > 0')
other_users = other_users.drop(columns=['na_sales', 'jp_sales', 'eu_sales', 'all_sales'])
other_users.columns = ['name', 
                     'platform', 
                     'year_of_release', 
                     'genre', 
                     'sales',
                     'critic_score', 
                     'user_score', 
                     'esrb_rating']
```

* **Средние пользовательские рейтинги жанров Action (англ. «действие», экшенигры) и Sports (англ. «спортивные соревнования») разные**.

Давайте начнем по порядку.

### [5.1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые](#content)

Сравним два массива данных:
* средние пользовательские рейтинги Xbox One;
* средние пользовательские рейтинги PC.

Давайте выдвенем нулевую и альтернативную гипотезу:

* **H0** - cредние пользовательские рейтинги платформ Xbox One и PC **одинаковые**;
* **H1** - cредние пользовательские рейтинги платформ Xbox One и PC **различаются**;

Упрощенная запись гипотезы:

* **X** - cредние пользовательские рейтинги платформ **Xbox One**;
* **P** - cредние пользовательские рейтинги платформ **PC**.

* **H0**: X = P;
* **H1**: X ≠ P.

Гипотеза двухсторонняя, потому что средние рейтинги могут отличатсяв большую и меньшую сторону.

пороговое значение **alpha = 5%**



```python
# проверим гипотезу №1 возьмем два массива и сравним их при помощи ttest,
# но перед этим надо понимать считать дисперсии равными или нет
# создадим два массива и выведим на экран их дисперсии
# если они отличаются более чем на 5% то в тесте будем использовать equal_var = False

# создадим 2 массива array

array_xb_one = games.query('year_of_release > 2012 and platform == "xone"')['user_score'].dropna()
array_xb_one = np.array(array_xb_one)

array_pc = games.query('year_of_release > 2012 and platform == "pc"')['user_score'].dropna()
array_pc = np.array(array_pc)

# сравним дисперсии и количество данных в этих массивам

print('''

Количчество данных в массиве Xbox One: {0};

Количество данных в массиве PC: {1}.


'''.format(len(array_xb_one), len(array_pc))
     )

print('''

дисперсия в массиве Xbox One: {0};

дисперсия данных в массиве PC: {1}.


'''.format(np.var(array_xb_one), np.var(array_pc))
     )
```

Несмотря на то, что в массивах данных количество не сильно отличается, их дисперсии слишком различны, поэтому при проверке гипотезы мы будем использовать **разные дисперсии**.

```python
# проверим гипотезу №1
# X - cредние пользовательские рейтинги платформ Xbox One;
# P - cредние пользовательские рейтинги платформ PC.
# H0: X = P;
# H1: X ≠ P.
# alpha = 0.05

alpha = .05

results = st.ttest_ind(array_xb_one, array_pc, equal_var = False)

print('р-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу. Cредние пользовательские рейтинги платформ Xbox One и PC различаются.')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Cредние пользовательские рейтинги платформ Xbox One и PC одинаковые.')




```

### [5.2: Средние пользовательские рейтинги жанров Action (англ. «действие», экшенигры) и Sports (англ. «спортивные соревнования») разные](#content)

Проведем гипотезу №2 по анологии с **первой** только на этот раз будем сравнивать пользовательские рейтинги по двум жанрам.

Сравним два массива данных:
* средние пользовательские рейтинги жанра Action;
* средние пользовательские рейтинги жанра Sports.

Давайте выдвенем нулевую и альтернативную гипотезу:

* **H0** - cредние пользовательские рейтинги жанров Action и Sports **одинаковые**;
* **H1** - cредние пользовательские рейтинги жанров Action и Sports **различаются**;

Упрощенная запись гипотезы:

* **A** - средние пользовательские рейтинги жанра **Action**;
* **S** - средние пользовательские рейтинги жанра **Sports**.

* **H0**: A = S;
* **H1**: A ≠ S.

Гипотеза двухсторонняя, потому что средние рейтинги могут отличатсяв большую и меньшую сторону.

пороговое значение **alpha = 5%**



```python
# все делаем по аналогии с первой гипотезой 

array_action = games.query('year_of_release > 2012 and genre == "action"')['user_score'].dropna()
array_action = np.array(array_action)

array_sports = games.query('year_of_release > 2012 and genre == "sports"')['user_score'].dropna()
array_sports = np.array(array_sports)

# сравним дисперсии и количество данных в этих массивам

print('''

Количчество данных в массиве Action: {0};

Количество данных в массиве Sports: {1}.


'''.format(len(array_action), len(array_sports))
     )

print('''

дисперсия в массиве Action: {0};

дисперсия данных в массиве Sports: {1}.


'''.format(np.var(array_action), np.var(array_sports))
     )
```

В массивах данных количество отличается, и их дисперсии слишком различны, поэтому при проверке гипотезы мы будем использовать **разные дисперсии**.

```python
# проверим гипотезу №2
# A - средние пользовательские рейтинги жанра Action;
# S - средние пользовательские рейтинги жанра Sports.
# H0: A = S;
# H1: A ≠ S.
# alpha = 0.5

alpha = .05

results = st.ttest_ind(array_action, array_sports, equal_var = False)

print('р-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports различаются.')
else:
    print('Не получилось отвергнуть нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports одинаковые.')

```
